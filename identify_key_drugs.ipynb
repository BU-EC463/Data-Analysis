```python
import json
import re

# Function to extract product names from the product description before the first number appears
def extract_product_name(description):
    # Search for the first occurrence of a number in the description
    match = re.search(r'\d', description)
    if match:
        # If a number is found, return the substring from the beginning to the position just before the number
        return description[:match.start()].strip()
    else:
        # If no number is found, return the whole description
        return description.strip()

# Create a dictionary to hold the primary ingredient as key and a list of associated product names as values
key_drugs = {}

# Iterate through the top 10 primary ingredients
for ingredient in top_ingredients_by_order_qty_manual.keys():
    # Filter the data for the current ingredient
    filtered_data = spreadsheet_data[spreadsheet_data['Primary Ingredient HIC4 Desc'] == ingredient]
    
    # Extract the product names for each of the filtered entries
    product_names = filtered_data['Product Description'].apply(extract_product_name)
    
    # Add the unique product names to the dictionary under the current ingredient
    key_drugs[ingredient] = list(product_names.unique())

# Save the dictionary to a JSON file
json_file_path = '/Users/bora/Downloads/identify_key_drugs.json'
with open(json_file_path, 'w') as json_file:
    json.dump(key_drugs, json_file)

# Display the path to the saved JSON file
json_file_path

with open(json_file_path, 'r') as json_file:
    key_drugs_data = json.load(json_file)

key_drugs_data
```

```python
# Function to extract package size from the product description, starting with the first number
def extract_package_size(description):
    # Search for the first occurrence of a number in the description
    match = re.search(r'\d', description)
    if match:
        # If a number is found, return the substring from the position of the number to the end
        return description[match.start():].strip()
    else:
        # If no number is found, return an empty string indicating no package size is described
        return ""

# Update the dictionary to include package sizes for each product name
for ingredient in key_drugs:
    # Create a list to hold tuples of product names and their package sizes
    product_names_and_sizes = []
    
    # Filter the data for the current ingredient
    filtered_data = spreadsheet_data[spreadsheet_data['Primary Ingredient HIC4 Desc'] == ingredient]
    
    # Iterate through each row of the filtered data
    for _, row in filtered_data.iterrows():
        # Extract the product name and package size
        product_name = extract_product_name(row['Product Description'])
        package_size = extract_package_size(row['Product Description'])
        # Append the tuple of product name and package size to the list
        product_names_and_sizes.append((product_name, package_size))
    
    # Remove duplicates by converting the list of tuples to a set, then back to a list
    product_names_and_sizes = list(set(product_names_and_sizes))
    
    # Sort the list of tuples alphabetically by product name
    product_names_and_sizes.sort(key=lambda x: x[0])
    
    # Update the dictionary with the list of product names and sizes
    key_drugs[ingredient] = product_names_and_sizes

# Save the updated dictionary to the JSON file
with open(json_file_path, 'w') as json_file:
    json.dump(key_drugs, json_file, indent=1)

# Display the updated dictionary
key_drugs
```

```python
# First, we need to create a function to extract the average price for each product.
def calculate_average_price(product_name, product_size):
    # Filter the data for entries that match the product name and size
    matching_products = spreadsheet_data[spreadsheet_data['Product Description'].str.contains(product_name) &
                                         spreadsheet_data['Product Description'].str.contains(product_size)]
    # Calculate the average price if there are any matching products
    if not matching_products.empty:
        return matching_products['AWP'].mean()
    else:
        # If no matching products are found, return None
        return None

# Update the dictionary to include average prices for each product name and package size
for ingredient, products in key_drugs.items():
    for i, (product_name, package_size) in enumerate(products):
        # Calculate the average price for each product
        average_price = calculate_average_price(product_name, package_size)
        # Update the tuple with the average price
        key_drugs[ingredient][i] = (product_name, package_size, average_price)

# Save the updated dictionary to the JSON file again
with open(json_file_path, 'w') as json_file:
    json.dump(key_drugs, json_file, indent=4)

# Display the updated dictionary with average prices
key_drugs

#HEPARIN, LIDOCAINE, ACETAMINOPHEN, CEFEPIME, CEFTRIAXONE, DAPTOMYCIN, INSULIN LISPRO, ALBUMIN HUMAN, ONDANSETRON, REMDESIVIR
```
